# Adult Census Income — 데이터 준비 EDA

Pandas / Polars 양쪽으로 로딩 결과를 비교하고, 결측치·중복을 처리한 뒤 기본 EDA를 수행한다.
로딩·클린 로직은 `src/data_loader.py` 에 재사용 가능한 함수로 분리되어 있다.

```python
import sys
import time

sys.path.insert(0, "..")
from src.data_loader import basic_eda, clean_pandas, load_pandas, load_polars
```

## Pandas vs Polars 로딩 비교

```python
t0 = time.perf_counter()
pdf_raw = load_pandas()
pandas_sec = time.perf_counter() - t0

t0 = time.perf_counter()
pldf = load_polars()
polars_sec = time.perf_counter() - t0

print(f"Pandas 로딩: {pandas_sec:.3f}s, shape={pdf_raw.shape}")
print(f"Polars 로딩: {polars_sec:.3f}s, shape={pldf.shape}")
```

## 결측치 · 중복 처리

```python
print("결측치(원본):")
print(pdf_raw.isna().sum()[pdf_raw.isna().sum() > 0])

pdf = clean_pandas(pdf_raw)
print(f"\n중복 제거 후 shape: {pdf.shape}")
```

## 기본 EDA

```python
basic_eda(pdf)
```

```python
pdf.to_parquet("../data/processed/adult_clean.parquet", index=False)
print("저장 완료: data/processed/adult_clean.parquet")
```
